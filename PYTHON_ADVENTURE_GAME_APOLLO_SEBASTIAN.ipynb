```python
# ============================================================================
# IMPORT LIBRARIES
# ============================================================================
import time  # For adding dramatic pauses in the game
import sys   # For system operations like exit
import os    # For clearing the screen
from typing import Dict, List, Tuple  # For type hints

# ============================================================================
# GLOBAL GAME STATISTICS
# ============================================================================
# These variables track player performance across the game session
game_stats = {
    'total_games': 0,
    'wins': 0,
    'losses': 0,
    'total_choices': 0,
    'paths_taken': [],
    'win_paths': [],
    'loss_paths': []
}
```

```python
# ============================================================================
# UTILITY FUNCTIONS
# ============================================================================

def clear_screen():
    """
    Clear the terminal screen for better readability.

    Purpose: Provides a clean interface between game sections
    Cross-platform: Works on Windows, Mac, and Linux
    """
    os.system('cls' if os.name == 'nt' else 'clear')


def print_slow(text: str, delay: float = 0.03):
    """
    Print text with a typewriter effect for dramatic impact.

    Args:
        text (str): The text to print
        delay (float): Delay between each character (default 0.03 seconds)

    Purpose: Creates an immersive storytelling experience
    GitHub Copilot Contribution: Suggested the character-by-character approach
    """
    for char in text:
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(delay)
    print()  # New line after text


def print_separator(char: str = "=", length: int = 70):
    """
    Print a decorative separator line.

    Args:
        char (str): Character to use for separator
        length (int): Length of the separator line

    Purpose: Visual organization of game sections
    """
    print(char * length)


def get_player_choice(options: List[str]) -> int:
    """
    Get and validate player input from a list of options.

    Args:
        options (List[str]): List of available choices

    Returns:
        int: The validated choice (1-based index)

    Purpose: Robust input handling with error checking
    GitHub Copilot Contribution: Suggested try-except error handling

    Example:
        options = ["Go left", "Go right"]
        choice = get_player_choice(options)
    """
    while True:
        try:
            choice = int(input("\nEnter your choice (number): "))
            if 1 <= choice <= len(options):
                game_stats['total_choices'] += 1
                return choice
            else:
                print(f"âŒ Please enter a number between 1 and {len(options)}")
        except ValueError:
            print("âŒ Invalid input! Please enter a number.")


def track_path(location: str, choice: str):
    """
    Track the player's journey through the game.

    Args:
        location (str): Current location in the game
        choice (str): Choice made by the player

    Purpose: Analytics for game balance and player behavior analysis
    """
    path_entry = f"{location}: {choice}"
    game_stats['paths_taken'].append(path_entry)


def display_game_stats():
    """
    Display comprehensive game statistics at the end.

    Purpose: Provide feedback on player performance
    Shows: Win rate, total games played, decision count
    """
    print_separator("=")
    print("ğŸ“Š GAME STATISTICS")
    print_separator("=")

    if game_stats['total_games'] > 0:
        win_rate = (game_stats['wins'] / game_stats['total_games']) * 100
        print(f"ğŸ® Total Games Played: {game_stats['total_games']}")
        print(f"ğŸ† Wins: {game_stats['wins']}")
        print(f"ğŸ’€ Losses: {game_stats['losses']}")
        print(f"ğŸ“ˆ Win Rate: {win_rate:.1f}%")
        print(f"ğŸ¯ Total Choices Made: {game_stats['total_choices']}")

        if game_stats['win_paths']:
            print(f"\nâœ… Successful Paths: {len(game_stats['win_paths'])}")

        if game_stats['loss_paths']:
            print(f"âŒ Failed Paths: {len(game_stats['loss_paths'])}")
    else:
        print("No games played yet!")

    print_separator("=")
```

```python
# ============================================================================
# TASK 3: CREATE THE FOREST PATH
# ============================================================================

def forest_path():
    """
    Handle the forest exploration scenario with multiple decision points.

    Purpose:
        - Present forest challenges
        - Offer strategic choices
        - Determine outcomes based on player decisions

    Decision Points:
        1. River crossing vs. Tree climbing
        2. Follow river upstream vs. Build a raft
        3. Wolf encounter - Fight vs. Befriend

    Outcomes:
        - Success: Find treasure map
        - Failure: Lost in forest or injured

    GitHub Copilot Contribution:
        - Suggested multiple branching paths
        - Recommended nested conditionals for depth
        - Proposed wildlife encounter mechanics

    TASK 3 OBSERVATION:
        This function demonstrates complex conditional logic with nested
        if-elif-else statements. It creates multiple narrative branches,
        showing how simple conditionals can create rich, interactive
        experiences. The use of track_path() enables analytics on player
        behavior. The function maintains narrative consistency while
        offering genuine player agency.
    """
    clear_screen()

    # Forest entrance scene
    print_separator("=")
    print_slow("\nğŸŒ² THE DARK FOREST")
    print_separator("=")
    print_slow("\nYou push through the thick undergrowth.")
    print_slow("Sunlight barely penetrates the dense canopy above.")
    print_slow("Every rustle makes you jump...")
    time.sleep(1)

    print_slow("\nAfter walking for what feels like hours, you hear rushing water.")
    print_slow("You've reached a wide river blocking your path.")
    time.sleep(0.5)

    # First decision point
    print_separator("-")
    print("\nğŸ’­ What will you do?")
    print("1. ğŸŠ Follow the river upstream (might find a crossing)")
    print("2. ğŸŒ³ Climb a tall tree (get a better view)")
    print("3. ğŸªµ Try to build a raft (cross the river)")

    choice1 = get_player_choice([1, 2, 3])
    track_path("Forest-River", f"Choice {choice1}")

    # BRANCH 1: Follow the river upstream
    if choice1 == 1:
        clear_screen()
        print_separator("=")
        print_slow("\nğŸï¸ RIVER EXPEDITION")
        print_separator("-")
        print_slow("\nYou follow the riverbank upstream.")
        print_slow("The sound of water is soothing...")
        time.sleep(1)

        print_slow("\nAfter an hour, you discover a narrow stone bridge!")
        print_slow("It looks ancient but sturdy.")
        print_slow("You cross safely to the other side.")
        time.sleep(1)

        print_slow("\nâš ï¸  But wait! A fierce wolf blocks your path!")
        print_slow("Its eyes glow in the dim forest light.")
        time.sleep(0.5)

        # Wolf encounter decision
        print_separator("-")
        print("\nğŸ’­ How will you handle this?")
        print("1. âš”ï¸  Draw your weapon and prepare to fight")
        print("2. ğŸ¥© Offer food to befriend the wolf")
        print("3. ğŸƒ Run back across the bridge")

        choice2 = get_player_choice([1, 2, 3])
        track_path("Forest-Wolf", f"Choice {choice2}")

        if choice2 == 1:
            # Fight the wolf - LOSS
            print_slow("\nâš”ï¸  You draw your weapon...")
            time.sleep(0.5)
            print_slow("The wolf lunges! You're not fast enough.")
            print_slow("You're injured and must retreat.")
            time.sleep(1)
            game_over(won=False, reason="Injured by wolf in combat")

        elif choice2 == 2:
            # Befriend the wolf - WIN PATH
            print_slow("\nğŸ¥© You slowly pull out some dried meat...")
            time.sleep(0.5)
            print_slow("The wolf approaches cautiously and takes the food.")
            print_slow("Its tail begins to wag!")
            time.sleep(1)
            print_slow("\nğŸº The wolf befriends you and leads you deeper into the forest!")
            print_slow("You follow your new companion...")
            time.sleep(1)
            forest_treasure_ending()

        else:
            # Run away - LOSS
            print_slow("\nğŸƒ You turn and run back across the bridge!")
            time.sleep(0.5)
            print_slow("The wolf chases you... you're getting tired...")
            print_slow("You stumble and fall into the river!")
            time.sleep(1)
            game_over(won=False, reason="Fell into the river while fleeing")

    # BRANCH 2: Climb a tree
    elif choice1 == 2:
        clear_screen()
        print_separator("=")
        print_slow("\nğŸŒ³ TREE CLIMBING")
        print_separator("-")
        print_slow("\nYou grab the lowest branch and start climbing.")
        print_slow("Higher and higher you go...")
        time.sleep(1)

        print_slow("\nFrom the top, you see:")
        print_slow("   ğŸ”ï¸  Mountains to the north")
        print_slow("   ğŸ›ï¸  Ruins to the east")
        print_slow("   ğŸŒŠ The river continues west")
        time.sleep(1)

        print_slow("\nâœ¨ You spot something glinting in the ruins!")
        time.sleep(0.5)

        # Ruins decision
        print_separator("-")
        print("\nğŸ’­ What's your next move?")
        print("1. ğŸ›ï¸  Head toward the ruins (might be treasure)")
        print("2. ğŸ”ï¸  Go to the mountains (safer route)")

        choice3 = get_player_choice([1, 2])
        track_path("Forest-TreeView", f"Choice {choice3}")

        if choice3 == 1:
            # Go to ruins - WIN PATH
            print_slow("\nğŸ›ï¸  You carefully climb down and head toward the ruins...")
            time.sleep(1)
            ruins_treasure_ending()

        else:
            # Go to mountains - LOSS
            print_slow("\nğŸ”ï¸  You head toward the mountains for safety...")
            time.sleep(0.5)
            print_slow("But the path is treacherous.")
            print_slow("A rockslide blocks your way!")
            print_slow("You're forced to turn back but night is falling...")
            time.sleep(1)
            game_over(won=False, reason="Lost in the mountains at nightfall")

    # BRANCH 3: Build a raft
    else:
        clear_screen()
        print_separator("=")
        print_slow("\nğŸªµ RAFT BUILDING")
        print_separator("-")
        print_slow("\nYou gather fallen branches and vines.")
        print_slow("You work for hours building a raft...")
        time.sleep(1)

        print_slow("\nâš ï¸  Your raft doesn't look very sturdy!")
        print_slow("The river current is strong...")
        time.sleep(0.5)

        # Raft decision
        print_separator("-")
        print("\nğŸ’­ Do you risk it?")
        print("1. ğŸš£ Use the raft (risky but direct)")
        print("2. ğŸš¶ Abandon the raft and walk upstream")

        choice4 = get_player_choice([1, 2])
        track_path("Forest-Raft", f"Choice {choice4}")

        if choice4 == 1:
            # Use raft - LOSS
            print_slow("\nğŸš£ You push the raft into the water and jump on...")
            time.sleep(0.5)
            print_slow("CRACK! The raft breaks apart!")
            print_slow("You're swept downstream by the current!")
            time.sleep(1)
            game_over(won=False, reason="Raft collapsed in the river")

        else:
            # Walk upstream - Alternative WIN PATH
            print_slow("\nğŸš¶ You decide against using the raft.")
            print_slow("Walking upstream, you find a safe crossing...")
            time.sleep(1)
            print_slow("\nOn the other side, you discover an old treasure map!")
            time.sleep(1)
            map_treasure_ending()


def forest_treasure_ending():
    """
    WIN scenario: Wolf leads to treasure.

    Purpose: Reward the player for choosing diplomacy over violence
    """
    clear_screen()
    print_separator("=")
    print_slow("\nğŸº THE WOLF'S SECRET")
    print_separator("=")
    print_slow("\nYour wolf companion leads you to a hidden grove.")
    print_slow("In the center stands an ancient stone pedestal...")
    time.sleep(1)

    print_slow("\nâœ¨ Upon the pedestal rests a glowing chest!")
    print_slow("You open it slowly...")
    time.sleep(1)

    print_slow("\nğŸ’ TREASURE FOUND!")
    print_slow("The chest is filled with gold, jewels, and ancient artifacts!")
    print_slow("The wolf sits beside you, wagging its tail.")
    time.sleep(1)

    game_over(won=True, reason="Befriended wolf and found the treasure!")


def ruins_treasure_ending():
    """
    WIN scenario: Ruins exploration successful.

    Purpose: Reward strategic thinking and exploration
    """
    clear_screen()
    print_separator("=")
    print_slow("\nğŸ›ï¸  THE ANCIENT RUINS")
    print_separator("=")
    print_slow("\nYou approach the crumbling stone structures.")
    print_slow("Vines cover most of the walls...")
    time.sleep(1)

    print_slow("\nYou spot an entrance partially hidden by vegetation.")
    print_slow("Inside, you find a chamber filled with ancient treasures!")
    time.sleep(1)

    print_slow("\nğŸ’ TREASURE FOUND!")
    print_slow("Golden coins, precious gems, and mysterious artifacts!")
    print_slow("Your strategic view from the tree paid off!")
    time.sleep(1)

    game_over(won=True, reason="Explored ruins and discovered the treasure!")


def map_treasure_ending():
    """
    WIN scenario: Found treasure map.

    Purpose: Reward patience and careful decision-making
    """
    clear_screen()
    print_separator("=")
    print_slow("\nğŸ—ºï¸  THE TREASURE MAP")
    print_separator("=")
    print_slow("\nYou examine the weathered map carefully.")
    print_slow("It shows an 'X' marking a location nearby!")
    time.sleep(1)

    print_slow("\nFollowing the map's directions, you dig at the marked spot...")
    print_slow("Your hands hit something solid!")
    time.sleep(1)

    print_slow("\nğŸ’ TREASURE FOUND!")
    print_slow("A buried chest filled with ancient riches!")
    print_slow("Your careful planning led to success!")
    time.sleep(1)

    game_over(won=True, reason="Found treasure map and located the treasure!")
```

```python
# ============================================================================
# TASK 4: CREATE THE CAVE PATH
# ============================================================================

def cave_path():
    """
    Handle the cave exploration scenario with darkness mechanics.

    Purpose:
        - Create atmospheric cave exploration
        - Implement light/darkness mechanics
        - Present unique cave challenges

    Decision Points:
        1. Light torch vs. Proceed in darkness
        2. Left tunnel vs. Right tunnel
        3. Swimming vs. Rock climbing

    Outcomes:
        - Success: Find underground treasure chamber
        - Failure: Lost in darkness or fall

    GitHub Copilot Contribution:
        - Suggested torch mechanics
        - Recommended environmental storytelling
        - Proposed multiple ending scenarios

    TASK 4 OBSERVATION:
        This function showcases how simple boolean logic (has_torch) can
        create branching narratives. The nested conditionals create depth
        while remaining readable. The atmospheric descriptions enhance
        player immersion. The function demonstrates good code organization
        with clear variable names and logical flow control.
    """
    clear_screen()

    # Cave entrance scene
    print_separator("=")
    print_slow("\nğŸ•³ï¸  THE MYSTERIOUS CAVE")
    print_separator("=")
    print_slow("\nYou stand at the cave entrance.")
    print_slow("Cold air flows from the darkness within.")
    print_slow("Water drips echo from deep inside...")
    time.sleep(1)

    print_slow("\nâš ï¸  It's very dark inside. You can barely see.")
    time.sleep(0.5)

    # First decision: Light source
    print_separator("-")
    print("\nğŸ’­ How will you proceed?")
    print("1. ğŸ”¦ Light a torch (safer but slower)")
    print("2. ğŸŒ‘ Proceed in darkness (faster but dangerous)")

    choice1 = get_player_choice([1, 2])
    track_path("Cave-Entrance", f"Choice {choice1}")

    # Store torch decision (boolean logic demonstration)
    has_torch = (choice1 == 1)

    # Process torch decision
    if has_torch:
        clear_screen()
        print_separator("=")
        print_slow("\nğŸ”¦ LIGHTING THE TORCH")
        print_separator("-")
        print_slow("\nYou strike your flint and steel.")
        print_slow("Sparks fly... the torch ignites!")
        print_slow("Warm light fills the cave entrance.")
        time.sleep(1)

        print_slow("\nâœ… You can see clearly now.")
        print_slow("Two tunnels branch ahead...")
        time.sleep(0.5)

    else:
        clear_screen()
        print_separator("=")
        print_slow("\nğŸŒ‘ INTO THE DARKNESS")
        print_separator("-")
        print_slow("\nYou step into the pitch-black cave.")
        print_slow("You can barely see your hand in front of your face.")
        print_slow("You proceed carefully, feeling along the walls...")
        time.sleep(1)

        print_slow("\nâš ï¸  You hear strange sounds ahead...")
        time.sleep(0.5)

    # Tunnel choice
    print_separator("-")

    if has_torch:
        print("\nğŸ’­ Which tunnel looks more promising?")
        print("1. â¬…ï¸  Left tunnel (you see stalactites)")
        print("2. â¡ï¸  Right tunnel (you hear water)")
    else:
        print("\nğŸ’­ You feel two openings... which way?")
        print("1. â¬…ï¸  Left opening (feels warmer)")
        print("2. â¡ï¸  Right opening (you hear water)")

    choice2 = get_player_choice([1, 2])
    track_path("Cave-Tunnels", f"Choice {choice2}")

    # LEFT TUNNEL PATH
    if choice2 == 1:
        clear_screen()
        print_separator("=")
        print_slow("\nâ¬…ï¸  LEFT TUNNEL")
        print_separator("-")

        if has_torch:
            print_slow("\nYour torch illuminates the narrow passage.")
            print_slow("Beautiful crystal formations line the walls!")
            time.sleep(1)

            print_slow("\nYou come to an underground lake.")
            print_slow("Across the water, you see a golden glow...")
            time.sleep(0.5)

            # Lake crossing with torch
            print_separator("-")
            print("\nğŸ’­ How will you cross?")
            print("1. ğŸŠ Swim across (torch will go out)")
            print("2. ğŸ§— Climb around on the rocks")

            choice3 = get_player_choice([1, 2])
            track_path("Cave-Lake-WithTorch", f"Choice {choice3}")

            if choice3 == 1:
                # Swim - LOSS (torch goes out)
                print_slow("\nğŸŠ You wade into the cold water...")
                time.sleep(0.5)
                print_slow("Your torch hisses and goes out!")
                print_slow("You're in complete darkness now.")
                print_slow("You lose your bearings and can't find the way...")
                time.sleep(1)
                game_over(won=False, reason="Lost in darkness after torch extinguished")

            else:
                # Climb - WIN
                print_slow("\nğŸ§— You carefully climb around the lake edge...")
                time.sleep(1)
                cave_treasure_ending_torch()

        else:
            # Left tunnel without torch - LOSS
            print_slow("\nYou feel your way through the darkness.")
            print_slow("The passage narrows...")
            time.sleep(0.5)
            print_slow("\nSuddenly, you feel nothing beneath your foot!")
            print_slow("You stumble and fall into a pit!")
            time.sleep(1)
            game_over(won=False, reason="Fell into a pit in the darkness")

    # RIGHT TUNNEL PATH
    else:
        clear_screen()
        print_separator("=")
        print_slow("\nâ¡ï¸  RIGHT TUNNEL")
        print_separator("-")

        if has_torch:
            print_slow("\nYour torch reveals an underground river.")
            print_slow("The water flows swiftly through a carved channel...")
            time.sleep(1)

            print_slow("\nâœ¨ On the far bank, you see something glittering!")
            print_slow("But the current is strong...")
            time.sleep(0.5)

            # River crossing decision
            print_separator("-")
            print("\nğŸ’­ What's your plan?")
            print("1. ğŸŠ Swim across with torch held high")
            print("2. ğŸš¶ Follow the river downstream")

            choice4 = get_player_choice([1, 2])
            track_path("Cave-River-WithTorch", f"Choice {choice4}")

            if choice4 == 1:
                # Swim with torch - WIN
                print_slow("\nğŸŠ You secure your torch and enter the water...")
                time.sleep(0.5)
                print_slow("The current is strong but you're a good swimmer!")
                print_slow("You make it across, torch still lit!")
                time.sleep(1)
                cave_treasure_ending_river()

            else:
                # Follow river - ALTERNATE WIN
                print_slow("\nğŸš¶ You decide to follow the river downstream...")
                time.sleep(1)
                print_slow("The passage opens into a massive cavern!")
                print_slow("Your torch reveals countless treasures scattered about!")
                time.sleep(1)
                cave_treasure_ending_cavern()

        else:
            # Right tunnel without torch - PARTIAL SUCCESS
            print_slow("\nYou feel rushing water nearby.")
            print_slow("The sound grows louder...")
            time.sleep(0.5)

            print_slow("\nâš ï¸  In the darkness, you accidentally step into the water!")
            print_slow("The current sweeps you away!")
            time.sleep(1)

            print_slow("\nğŸŒŠ You're tumbling through water in complete darkness...")
            time.sleep(0.5)
            print_slow("Suddenly, you're spit out into daylight!")
            print_slow("You're outside the cave, soaking wet.")
            time.sleep(1)

            print_slow("\nğŸ˜… You survived, but found no treasure.")
            print_slow("Sometimes it's better to have light in the darkness...")
            time.sleep(1)

            game_over(won=False, reason="Swept out of cave by underground river")


def cave_treasure_ending_torch():
    """
    WIN scenario: Cave treasure with torch navigation.

    Purpose: Reward careful planning and resource management
    """
    clear_screen()
    print_separator("=")
    print_slow("\nğŸ’¡ THE ILLUMINATED PATH")
    print_separator("=")
    print_slow("\nYou reach the source of the golden glow.")
    print_slow("Your torch reveals a hidden chamber!")
    time.sleep(1)

    print_slow("\nâœ¨ Ancient treasure chests line the walls!")
    print_slow("Gold coins spill across the floor.")
    print_slow("Precious gems sparkle in your torchlight.")
    time.sleep(1)

    print_slow("\nğŸ’ TREASURE FOUND!")
    print_slow("Your decision to bring light was wise!")
    print_slow("The torch revealed the path to riches!")
    time.sleep(1)

    game_over(won=True, reason="Used torch to safely navigate to treasure!")


def cave_treasure_ending_river():
    """
    WIN scenario: Successfully crossed the river.

    Purpose: Reward bravery and calculated risk-taking
    """
    clear_screen()
    print_separator("=")
    print_slow("\nğŸŠ THE RIVER CROSSING")
    print_separator("=")
    print_slow("\nYou reach the far bank, dripping but triumphant.")
    print_slow("Before you lies a treasure cache!")
    time.sleep(1)

    print_slow("\nğŸ’° Ancient coins and artifacts!")
    print_slow("The glittering you saw was real gold!")
    time.sleep(1)

    print_slow("\nğŸ’ TREASURE FOUND!")
    print_slow("Your swimming skills paid off!")
    print_slow("Fortune favors the brave!")
    time.sleep(1)

    game_over(won=True, reason="Swam across river to reach treasure!")


def cave_treasure_ending_cavern():
    """
    WIN scenario: Discovered the main treasure cavern.

    Purpose: Reward exploration and patience
    """
    clear_screen()
    print_separator("=")
    print_slow("\nğŸ›ï¸  THE TREASURE CAVERN")
    print_separator("=")
    print_slow("\nYour torch illuminates a breathtaking sight.")
    print_slow("This cavern is filled with treasures from countless ages!")
    time.sleep(1)

    print_slow("\nâœ¨ Gold stacked in piles!")
    print_slow("ğŸ’ Gems scattered like pebbles!")
    print_slow("ğŸ‘‘ Ancient crowns and artifacts!")
    time.sleep(1)

    print_slow("\nğŸ† JACKPOT! You found the MAIN treasure chamber!")
    print_slow("Following the river led you to the greatest treasure!")
    time.sleep(1)

    game_over(won=True, reason="Discovered the legendary main treasure chamber!")
```

```python
# ============================================================================
# GAME ENDING FUNCTIONS
# ============================================================================

def game_over(won: bool, reason: str):
    """
    Handle game ending and offer replay option.

    Args:
        won (bool): Whether the player won or lost
        reason (str): Description of how the game ended

    Purpose:
        - Display appropriate ending message
        - Update game statistics
        - Offer replay option

    Flow:
        1. Show win/loss message
        2. Update statistics
        3. Display current path taken
        4. Offer to replay or quit
    """
    print_separator("=")

    if won:
        # Victory message
        print_slow("\nğŸ‰ CONGRATULATIONS! ğŸ‰")
        print_separator("-")
        print_slow(f"\nâœ… {current_player_name}, you have succeeded!")
        print_slow(f"ğŸ’ {reason}")
        print_slow("\nYour name will be remembered in legends!")

        # Update win statistics
        game_stats['wins'] += 1
        game_stats['win_paths'].append(game_stats['paths_taken'].copy())

    else:
        # Loss message
        print_slow("\nğŸ’€ GAME OVER ğŸ’€")
        print_separator("-")
        print_slow(f"\nâŒ {current_player_name}, your adventure has ended.")
        print_slow(f"âš ï¸  {reason}")
        print_slow("\nBut every failure is a lesson learned...")

        # Update loss statistics
        game_stats['losses'] += 1
        game_stats['loss_paths'].append(game_stats['paths_taken'].copy())

    # Show path taken this game
    print_separator("-")
    print("\nğŸ“ YOUR JOURNEY:")
    for i, step in enumerate(game_stats['paths_taken'], 1):
        print(f"   {i}. {step}")

    # Clear path for next game
    game_stats['paths_taken'] = []

    time.sleep(2)

    # Offer replay
    print_separator("=")
    print("\nğŸ”„ Would you like to play again?")
    print("1. Yes - Start a new adventure")
    print("2. No - View statistics and exit")

    replay_choice = get_player_choice([1, 2])

    if replay_choice == 1:
        # Restart the game
        print_slow("\nğŸ”„ Starting new adventure...")
        time.sleep(1)
        start_game()
    else:
        # Show statistics and exit
        print_slow("\nğŸ‘‹ Thank you for playing!")
        time.sleep(0.5)
        display_game_stats()
        print_separator("=")
        print_slow("\nMay your future adventures be legendary! âš”ï¸")
        print_separator("=")
        sys.exit(0)


```
